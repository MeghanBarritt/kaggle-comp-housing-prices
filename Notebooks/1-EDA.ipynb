Train correlation coefficients of note (>0.5), rounded to 4 decimal places.

```python
print(f'Sale Price + Overall Quality: {round(np.corrcoef(train_df['SalePrice'],train_df['OverallQual'])[0][1],4)}')
print(f'Sale Price + Year Built: {round(np.corrcoef(train_df['SalePrice'],train_df['YearBuilt'])[0][1],4)}')
print(f'Sale Price + Year Remodelled: {round(np.corrcoef(train_df['SalePrice'],train_df['YearRemodAdd'])[0][1],4)}')
print(f'Sale Price + Total Basement Area: {round(np.corrcoef(train_df['SalePrice'],train_df['TotalBsmtSF'])[0][1],4)}')
print(f'Sale Price + 1st Floor Area: {round(np.corrcoef(train_df['SalePrice'],train_df['1stFlrSF'])[0][1],4)}')
print(f'Sale Price + Above Ground Area: {round(np.corrcoef(train_df['SalePrice'],train_df['GrLivArea'])[0][1],4)}')
print(f'Sale Price + Full Bath: {round(np.corrcoef(train_df['SalePrice'],train_df['FullBath'])[0][1],4)}')
print(f'Sale Price + Rooms Above Ground: {round(np.corrcoef(train_df['SalePrice'],train_df['TotRmsAbvGrd'])[0][1],4)}')
print(f'Sale Price + Garage Car Capacity: {round(np.corrcoef(train_df['SalePrice'],train_df['GarageCars'])[0][1],4)}')
print(f'Sale Price + Garage Area: {round(np.corrcoef(train_df['SalePrice'],train_df['GarageArea'])[0][1],4)}')
```

```python
y = train_df['SalePrice']
a = train_df['OverallQual']
b = train_df['YearBuilt']
c = train_df['YearRemodAdd']
d = train_df['TotalBsmtSF']
e = train_df['1stFlrSF']
f = train_df['GrLivArea']
g = train_df['FullBath']
h = train_df['TotRmsAbvGrd']
i = train_df['GarageCars']
j = train_df['GarageArea']
```

```python
aa, ay = np.polyfit(a,y,1)
bb, by = np.polyfit(b,y,1)
cc, cy = np.polyfit(c,y,1)
dd, dy = np.polyfit(d,y,1)
ee, ey = np.polyfit(e,y,1)
ff, fy = np.polyfit(f,y,1)
gg, gy = np.polyfit(g,y,1)
hh, hy = np.polyfit(h,y,1)
ii, iy = np.polyfit(i,y,1)
jj, jy = np.polyfit(j,y,1)
```

```python
plt.figure(figsize=(10,15))
plt.subplot(5,2,1)
plt.scatter(a,y)
plt.plot(a, aa*a+ay,color='red')
plt.xlabel('OverallQual')
plt.ylabel('SalePrice')

plt.subplot(5,2,2)
plt.scatter(b,y)
plt.plot(b, bb*b+by,color='red')
plt.xlabel('YearBuilt')
plt.ylabel('SalePrice')

plt.subplot(5,2,3)
plt.scatter(c,y)
plt.plot(c, cc*c+cy,color='red')
plt.xlabel('YearRemodAdd')
plt.ylabel('SalePrice')

plt.subplot(5,2,4)
plt.scatter(d,y)
plt.plot(d, dd*d+dy,color='red')
plt.xlabel('TotalBsmtSF')
plt.ylabel('SalePrice')

plt.subplot(5,2,5)
plt.scatter(e,y)
plt.plot(e, ee*e+ey,color='red')
plt.xlabel('1stFlrSF')
plt.ylabel('SalePrice')

plt.subplot(5,2,6)
plt.scatter(f,y)
plt.plot(f, ff*f+fy,color='red')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')

plt.subplot(5,2,7)
plt.scatter(g,y)
plt.plot(g, gg*g+gy,color='red')
plt.xlabel('FullBath')
plt.ylabel('SalePrice')

plt.subplot(5,2,8)
plt.scatter(h,y)
plt.plot(h, hh*h+hy,color='red')
plt.xlabel('TotRmsAbvGrd')
plt.ylabel('SalePrice')

plt.subplot(5,2,9)
plt.scatter(i,y)
plt.plot(i, ii*i+iy,color='red')
plt.xlabel('GarageCars')
plt.ylabel('SalePrice')

plt.subplot(5,2,10)
plt.scatter(j,y)
plt.plot(j, jj*j+jy,color='red')
plt.xlabel('GarageArea')
plt.ylabel('SalePrice')

plt.suptitle('Train data correlations\n')
plt.tight_layout()
```
